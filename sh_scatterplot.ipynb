```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_clean_airbnb.csv')
df
```

```python
df = df[['price', 'number_of_reviews', 'reviews_per_month']]
print(df.describe())

# Filter in order to better see the relationship, only looking where Price < 1000 and Number of Reviews > 0. 
df = df[(df['price'] < 600) & (df['number_of_reviews'] > 0)]
sns.regplot(x=df['number_of_reviews'], y=df['price'], ci=False, line_kws={'color':'red'})
plt.xlabel('Number of Reviews') 
plt.ylabel('Price')
plt.title('Price vs. Number of Reviews')
```
